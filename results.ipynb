#### Get Mean and Std dataframes

```python
#averages and std deviation
df = deepcopy(df_original)
# identify the columns we want to aggregate by; this could
# equivalently be defined as list(df.columns[0:3])
group_cols = ['task_name']

# identify the columns which
# we want to average; this could
# equivalently be defined as list(df.columns[4:])
metric_cols = list(df.columns[2:10])
# create a new DataFrame with a MultiIndex consisting of the group_cols
# and a column for the mean of each column in metric_cols
aggs_mean = df.groupby(group_cols)[metric_cols].mean()
aggs_std = df.groupby(group_cols)[metric_cols].std()
#df_to_merge =  df.merge(right=aggs_mean, right_index=True, left_on=group_cols, how='right')
#print((aggs_mean))

# remove the metric_cols from df because we are going to replace them
# with the means in aggs
df.drop(metric_cols, axis=1, inplace=True)
# dedupe to leave only one row with each combination of group_cols
# in df
df.drop_duplicates(subset=group_cols, keep='last', inplace=True)
# add the mean columns from aggs into df
df_mean = df.merge(right= aggs_mean, right_index=True, left_on=group_cols, how='right')
df_mean = df_mean.drop(columns=["task_name_with_run"])

df_std = df.merge(right= aggs_std, right_index=True, left_on=group_cols, how='right')
df_std = df_std.drop(columns=["task_name_with_run"])
```
